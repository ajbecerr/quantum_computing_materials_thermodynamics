```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#units kJ/mol
DFT_CO2_Ru = 0.0
DFT_CO_O_TS_Ru = 51.646032
DFT_CO_O_Ru = 23.109867
```

```python
#arbitrary units coarse model
CO2_Ru = -0.01360945
CO_O_TS_Ru = -0.01360521
CO_O_Ru = -0.01360595

reference = CO2_Ru

CO2_Ru = CO2_Ru - reference
CO_O_TS_Ru = CO_O_TS_Ru - reference
CO_O_Ru = CO_O_Ru - reference
```

```python
plt.plot([0,1,2],[DFT_CO2_Ru,DFT_CO_O_TS_Ru,DFT_CO_O_Ru])
plt.ylabel('E (kj/mol)')
plt.xticks([0, 1, 2])
plt.savefig('DFT_CO2_break.png',dpi=300)
```

```python
plt.plot([0,1,2],[CO2_Ru,CO_O_TS_Ru,CO_O_Ru])
plt.ylabel('E arbitrary units')
plt.xticks([0, 1, 2])
plt.savefig('coarse_CO2_break.png',dpi=300)
```
