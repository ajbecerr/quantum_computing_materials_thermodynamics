```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#units kJ/mol
DFT_CO2_Ru = 0.0
DFT_CO_O_TS_Ru = 51.646032
DFT_CO_O_Ru = 23.109867
```

```python
#arbitrary units coarse model
CO2_Ru = -0.01360945
CO_O_TS_Ru = -0.01360521
CO_O_Ru = -0.01360595

reference = CO2_Ru

CO2_Ru = CO2_Ru - reference
CO_O_TS_Ru = CO_O_TS_Ru - reference
CO_O_Ru = CO_O_Ru - reference
```

```python
# quantum simulator weighted mean approach
#arbitrary units coarse model
CO2_Ru_sim = -0.0001220703125
CO_O_TS_Ru_sim = 0.00042724609375
CO_O_Ru_sim = -0.00164794921875

reference_sim = CO2_Ru_sim

CO2_Ru_sim = CO2_Ru_sim - reference_sim
CO_O_TS_Ru_sim = CO_O_TS_Ru_sim - reference_sim
CO_O_Ru_sim = CO_O_Ru_sim - reference_sim
```

```python
#Changing eigenvalue method- the only peak that changed went 498-500-499
CO2_Ru_change = 498
CO_O_TS_Ru_change = 500
CO_O_Ru_change = 499
```

```python
# The highest peak for 8 qubits
divisor = np.power(2,10)
print(np.power(2,10))
CO2_Ru_fine = (17*int('0001110111',2) + 1625*int('0001111000',2) + 11*int('0001111001',2))/divisor
CO_O_TS_Ru_fine = (23*int('0001110111',2) + 1619*int('0001111000',2) + 25*int('0001111001',2))/divisor
CO_O_Ru_fine = (39*int('0001110111',2) + 1550*int('0001111000',2) + 12*int('0001111001',2))/divisor

reference_fine = CO2_Ru_fine
CO2_Ru_fine = CO2_Ru_fine - reference_fine
CO_O_TS_Ru_fine = CO_O_TS_Ru_fine - reference_fine
CO_O_Ru_fine = CO_O_Ru_fine - reference_fine
```

```python
plt.plot([0,1,2],[DFT_CO2_Ru,DFT_CO_O_TS_Ru,DFT_CO_O_Ru])
plt.ylabel('E (kj/mol)')
plt.xticks([0, 1, 2])
plt.savefig('DFT_CO2_break.png',dpi=300)
```

```python
plt.plot([0,1,2],[CO2_Ru,CO_O_TS_Ru,CO_O_Ru])
plt.ylabel('E arbitrary units')
plt.xticks([0, 1, 2])
plt.savefig('coarse_CO2_break.png',dpi=300)
```

```python
plt.plot([0,1,2],[CO2_Ru_sim,CO_O_TS_Ru_sim,CO_O_Ru_sim])
plt.ylabel('E arbitrary units')
plt.xticks([0, 1, 2])
plt.savefig('quantum_sim_CO2_break.png',dpi=300,bbox_inches='tight')
```

```python
plt.plot([0,1,2],[CO2_Ru_fine,CO_O_TS_Ru_fine,CO_O_Ru_fine])
plt.ylabel('E arbitrary units')
plt.xticks([0, 1, 2])
plt.savefig('fine_sim_CO2_break.png',dpi=300,bbox_inches='tight')
```

```python
plt.plot([0,1,2],[CO2_Ru_change,CO_O_TS_Ru_change,CO_O_Ru_change])
plt.ylabel('E arbitrary units')
plt.xticks([0, 1, 2])
plt.savefig('change_CO2_break.png',dpi=300,bbox_inches='tight')
```
